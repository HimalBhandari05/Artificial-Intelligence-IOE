```python
from pprint import pprint

n = 9
board = [['_']*n for i in range (n)]
pprint(board)
```

```python
# PLACING THE QWEEN ON ROW 1 COLUMN 1 

board[0][0] = 'Q'
pprint(board)
```

```python
def is_safe(board, row , col, n):
    # Column Test
    for x in range(row):
        if board[x][col] == 'Q':
            return False
    
    # Diagonal Test

    # left diagonal test.
    r = row -1 
    c = col -1 
    while (r>=0 and c >=0):
        if board[r][c] == "Q":
            return False
        r = r-1
        c = c-1

    # Top right diagonal test.

    r = row - 1
    c = col + 1

    while( r >= 0 and c <n ):
        if board[r][c] == "Q":
            return False
        r = r - 1
        c = c + 1

    return True
```

```python
pprint(is_safe(board , 1, 0 , n))
```

```python
def nQueensSOlver(board , x , n):
    if x>=n:
        return True
    for col in range(n):
        if (is_safe(board , x , col , n)):
            board[x][col] = 'Q'
            if (nQueensSOlver(board , x+1, n)):
                return True
            board[x][col] = "_"
        
    return False
```
