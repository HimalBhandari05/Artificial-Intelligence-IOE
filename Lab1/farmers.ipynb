```python
from pprint import pprint

Initial = (0, 0, 0, 0)
Final = (1, 1, 1, 1)
```

```python
def is_safe(state):
    farmer, cabbage, wolf, goat = state

    # Wolf ra goat sangei hunu vayena.
    if wolf == goat and farmer != goat:
        return False

    # cabbage ra goat sangei hunu vayena if farmer absent vaye.
    if cabbage == goat and farmer != goat:
        return False

    return True
```

```python
def neighbours(state):
    f, c, w, g = state
    steps = []

    new = (1 - f, c, w, g)
    if is_safe(new):
        steps.append((new, "Farmer alone"))

    # farmer ra wolf sangei vayo vaney :( goat ra cabbage sangei hunxan ni ta.)
    if f == w:
        new = (1 - f, c, 1 - w, g)
        if is_safe(new):
            steps.append((new, "Farmer takes Wolf"))

    # farmer ra cabbage sangei vayo vaney: ( goat ra wolf sangei hunxa ni ta.)
    if f == c:
        new = (1 - f, 1 - c, w, g)
        if is_safe(new):
            steps.append((new, "Farmer takes Cabbage"))

    if f == g:
        new = (1 - f, c, w, 1 - g)
        if is_safe(new):
            steps.append((new, "Farmer takes Goat"))

    return steps
```

```python
def dfs(start, goal):
    stack = [start]
    visited = set()
    parent = {}

    while stack:
        state = stack.pop()

        if state == goal:
            print("Goal Found!\n")

            path = []
            current = goal

            while current in parent:
                prev, move = parent[current]
                path.append((move, current))
                current = prev

            path.reverse()

            print("Solution Path:")
            print(start)
            for move, step in path:
                print(move)
                print(step)

            return path

        if state in visited:
            continue

        visited.add(state)

        for nxt, move in neighbours(state):
            if nxt not in visited:
                parent[nxt] = (state, move)
                stack.append(nxt)

    print("No Solution Found.")
    return None
```

```python
path = dfs(Initial, Final)
```
